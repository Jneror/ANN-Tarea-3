# 1) Reducción de dimensionalidad
Para esta primera sección, gracias a la simplicidad del problema tratado, se experimentará con un autoencoder tradicional (*feed forward*) en donde las capas de éste sean densas. Para esto se re estructurarán los datos de entradas en forma de vector, es decir la matriz de 28 $\times$ 28 pasa a ser un vector de 784 componentes.

```python
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
x_val = x_val.reshape((len(x_val), np.prod(x_val.shape[1:])))

print("Dimensión x_train: ", x_train.shape)
print("Dimensión x_test:", x_test.shape)
print("Dimensión x_val: ",x_val.shape)
```

Una de las aplicaciones tı́picas de un AE es reducción de dimensionalidad, es decir, implementar una transformación $\phi:{\rm I\!R}^d \rightarrow {\rm I\!R}^{d'}$ de objetos representados originalmente por $d$ atributos en una nueva representación de $d'$ atributos, de modo tal que se preserve lo mejor posible la “información” original. Obtener tal representación es útil desde un punto de vista computacional (compresión) y estadı́stico (permite construir modelos con un menor número de parámetros libres). Un AE es una técnica de reducción de dimensionalidad no supervisada porque no hace uso de información acerca de las clases a las que pertenecen los datos de entrenamiento.

Binary crossentropy es la función de error a utlizar debido a que los pixeles que no son ceros ni unos es contorno y ruido. Pixeles prendidos o apagados son los que tienen información más ulti, por lo que se elige esta funcion que toma enfasis en esta clase de puntos.

 Esta es una tarea de clasificación de múltiples etiquetas, no una clasificación de múltiples clases: en el caso anterior, cada etiqueta corresponde a un problema de clasificación binaria, un comentario podría tener varias etiquetas; en este último caso, un comentario solo se puede asignar como una etiqueta.

```python
def iterateOverDim(activEncod, activDecod, label):
    
    for encoding_dim in [2,8,32,64]:
        
        #Autoencoder
        input_img = Input(shape=(784,))
        encoded = Dense(encoding_dim, activation = activEncod)(input_img)
        decoded = Dense(784, activation = activDecod)(encoded)
        autoencoder = Model(input_img, decoded)
        autoencoder.summary()
        
        #Encoder
        input_img = Input(shape=(784,))
        encoder_layer = autoencoder.layers[-2]
        encoder = Model(input_img, encoder_layer(input_img))
        encoder.summary()
        
        #Decoder
        encoded_input = Input(shape = (encoding_dim,))
        decoder_layer = autoencoder.layers[-1]
        decoder = Model(encoded_input, decoder_layer(encoded_input))
        decoder.summary()
        
        #Compile and fit
        autoencoder.compile(optimizer=SGD(lr = 1.0), loss='binary_crossentropy', metrics = ["acc"])
        pd.DataFrame(autoencoder.fit(x_train, x_train, epochs=50, batch_size=32,validation_data=(x_val, x_val)).history).to_csv("history_2A_"+label+"_"+str(encoding_dim)+".csv")
        
        #Save
        autoencoder.save("autoencoder_2A_"+label+"_"+str(encoding_dim)+".h5")
        encoder.save("encoder_2A_"+label+"_"+str(encoding_dim)+".h5")
        decoder.save("decoder_2A_"+label+"_"+str(encoding_dim)+".h5")
```

## Pregunta C
Para verificar la calidad de la representación obtenida, implemente el clasificador denominado $kNN$ (k-nearest neighbor): dada una imagen $x$, el clasificador busca las k = 10 imágenes de entrenamiento más similares (de acuerdo a una distancia, e.g. euclidiana) y predice como clase, la etiqueta más popular entre las imágenes cercanas. **Mida el error de pruebas** obtenido construyendo este clasificador sobre la data reducida a través del *autoencoder* comparando con la representación reducida obtenida vía PCA (una técnica clásica de reducción de dimensionalidad) utilizando el mismo número de dimensiones $d'$= 2, 4, 8, 16, 32. Considere tanto el error de reconstrucción como el desempeño en clasificación , además de comparar los tiempos medios de predicción en ambos escenarios **¿La representación generada por el *autoencoder* logra generalizar?**

```python
import timeit
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier

encoderTimes = []
pcaTimes = []
encoderAcc = []
pcaAcc = []

for encoding_dim in [2,8,32,64]:
    print(encoding_dim)
    #Encoder
    autoencoder = load_model("autoencoder_2A_relusig_"+str(encoding_dim)+".h5")
    input_img = Input(shape=(784,))
    encoder_layer = autoencoder.layers[-2]
    encoder = Model(input_img, encoder_layer(input_img))
    
    start = timeit.timeit()
    encoded_train = encoder.predict(x_train)
    encoderTimes.append(timeit.timeit() - start)
    
    start = timeit.timeit()
    encoded_test = encoder.predict(x_test)
    encoderTimes.append(timeit.timeit() - start)
    
    #PCA
    pca = PCA(n_components = encoding_dim)
    pca.fit(x_train)
    
    
    start = timeit.timeit()
    pca_train = pca.transform(x_train)
    pcaTimes.append(timeit.timeit() - start)
    
    start = timeit.timeit()
    pca_test = pca.transform(x_test)
    pcaTimes.append(timeit.timeit() - start)
    
    #Classification
    clf = KNeighborsClassifier(10)
    clf.fit(pca_train, y_train)
    pcaAcc.append(clf.score(pca_test,y_test))
    
    clf = KNeighborsClassifier(10)
    clf.fit(encoded_train, y_train)
    encoderAcc.append(clf.score(encoded_test,y_test))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].plot([2,8,32,64], encoderAcc, color = pink, label = "Encoder")
ax[0].plot([2,8,32,64], pcaAcc, color = green, label = "PCA")
ax[0].legend()
ax[0].set_xlabel("Encoding dimension")
ax[0].set_ylabel("Accuracy")
ax[0].set_title("Accuracy vs Encoding dimension")

ax[1].plot(encoderTimes, color = pink, label = "Encoder")
ax[1].plot(pcaTimes, color = green, label = "PCA")
ax[1].legend()
ax[1].set_xlabel("nth time")
ax[1].set_ylabel("Time [s]")
ax[1].set_title("Prediction time")
plt.show()
```

Se puede observar de los gráficos que con PCA se obtienen mejor resultados que con el encoder, pero no están muy alejados. Con relación a los tiempos, la predicción del encoder es más rápida.

## Pregunta B
 Construya y entrene una red neuronal profunda para clasificar las imágenes de MNIST utilizando la arquitectura propuesta en (a) y pre-entrenando los pesos de cada capa mediante un autoencoder básico. Proceda en modo clásico, es decir, entrenando en modo no supervisado una capa a la vez y tomando como input de cada nivel la representación (entrenada) obtenida en el nivel anterior. Después del entrenamiento efectúe un entrenamiento supervisado convencional (*fine-tunning*). **Compare los resultados de clasificación sobre el conjunto de pruebas con aquellos obtenidos en (a), sin pre-entrenamiento. Evalúe también los resultados antes del *fine-tunning*.** Comente.

Ahora vamos a entrenar una red deep, pero primero vamos a entregarle los pesos ya pre-entrenados por un autoencoder. Es decir, se realiza  en modo clásico, es decir, entrenando en modo no supervisado una capa a la vez y tomando como input de cada nivel la representación (entrenada) obtenida en el nivel anterior. 

```python
from keras.datasets import mnist
... ## Load and preprocess MNIST as usual
(x_train, y_train), (x_test, y_test) = mnist.load_data()
total=len(x_test)
x_train = x_train.astype('float32') / 255.
x_test= x_test.astype('float32') / 255.
x_testt=x_test[:int(0.5*total)]                       
x_val=x_test[int(0.5*total)::]  
y_val=y_test[int(0.5*total)::]
y_test=y_test[:int(0.5*total)]  

x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_testt.reshape((len(x_testt), np.prod(x_testt.shape[1:])))
x_val = x_val.reshape((len(x_val), np.prod(x_val.shape[1:])))


Y_train = to_categorical(y_train, 10)
Y_test = to_categorical(y_test, 10)
y_val =to_categorical(y_val,10)
```

```python
...###AUTOENCODER 1

activation_layer1 = 'sigmoid'
decoder_activation_1 = 'sigmoid'
epochs= 40
n_hidden_layer1 =64
batch_size_ = 32
loss_= 'binary_crossentropy'
input_img1 = Input(shape=(784,))
encoded1 = Dense(n_hidden_layer1,activation=activation_layer1)(input_img1)
decoded1 = Dense(784, activation=decoder_activation_1)(encoded1)
autoencoder1 = Model(inputs=input_img1, outputs=decoded1)
encoder1 = Model(inputs=input_img1, outputs=encoded1)
autoencoder1.compile(optimizer=optimizer_, loss=loss_)
autoencoder1.fit(x_train, x_train, nb_epoch=epochs_, batch_size=batch_size_,shuffle=True, validation_data=(x_val, x_val))
encoded_input1 = Input(shape=(n_hidden_layer1,))
autoencoder1.save('autoencoder_layer1.h5')
encoder1.save('encoder_layer1.h5')
```

```python
...###AUTOENCODER 2
activation_layer2 = 'sigmoid'
decoder_activation_2 = 'sigmoid'
epochs= 40
n_hidden_layer1 =64
batch_size_ = 32
loss_= 'binary_crossentropy'

x_train_encoded1 = encoder1.predict(x_train) #FORWARD PASS DATA THROUGH FIRST ENCODER
x_val_encoded1 = encoder1.predict(x_val)
x_test_encoded1 = encoder1.predict(x_test)
input_img2 = Input(shape=(n_hidden_layer1,))
encoded2 = Dense(n_hidden_layer2, activation=activation_layer2)(input_img2)
decoded2 = Dense(n_hidden_layer2, activation=decoder_activation_2)(encoded2)
autoencoder2 = Model(inputs=input_img2, outputs=decoded2)
encoder2 = Model(inputs=input_img2, outputs=encoded2)
autoencoder2.compile(optimizer=optimizer_, loss=loss_)
autoencoder2.fit(x_train_encoded1,x_train_encoded1,nb_epoch=epochs_,batch_size=batch_size_,shuffle=True, validation_data=(x_val_encoded1, x_val_encoded1))
encoded_input2 = Input(shape=(n_hidden_layer2,))
autoencoder2.save('autoencoder_layer2.h5')
encoder2.save('encoder_layer2.h5')
```

Aquí, se realiza el aprendizaje supervisado (fine-tunning):

```python
...#FINE TUNNING
from keras.models import Sequential
model = Sequential()
model.add(Dense(n_hidden_layer1, activation=activation_layer1, input_shape=(784,)))
model.layers[-1].set_weights(autoencoder1.layers[1].get_weights())
model.add(Dense(n_hidden_layer2, activation=activation_layer2))
model.layers[-1].set_weights(autoencoder2.layers[1].get_weights())
model.add(Dense(10, activation='softmax'))
model.summary()
model.compile(optimizer=optimizer_,loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, Y_train,nb_epoch=20, batch_size=25,
shuffle=True, validation_data=(x_val, Y_val))
model.save('Net-768x1000x1000x10-finetunned.h5')
```
